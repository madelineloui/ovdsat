## Few-shot Plots

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

def plot_fewshot(data, title):

    # X axis
    shots = ["0", "1", "3", "5", "10", "30"]
    x = np.array([0, 1, 3, 5, 10, 30])

    # Backbone grouping
    color_groups = {
        "CLIP": ["CLIP", "CLIP + CoOp"],
        "GeoRSCLIP": ["GeoRSCLIP", "GeoRSCLIP + CoOp"],
        "RemoteCLIP": ["RemoteCLIP", "RemoteCLIP + CoOp"],
        #"RemoteCLIP + fmow-mm": ["RemoteCLIP + fmow-mm", "RemoteCLIP + fmow-mm + CoOp"],
        #"GeoRSCLIP + fmow-mm": ["GeoRSCLIP + fmow-mm", "GeoRSCLIP + fmow-mm + CoOp"],
        "DINOv2": ["DINOv2"],
    }

    # Distinct color assignment (auto, high contrast)
    cmap = plt.get_cmap("tab10")

    color_map = {}
    for i, base_name in enumerate(color_groups.keys()):
        color_map[base_name] = cmap(i)

    # Plot
    plt.figure(figsize=(9, 7))   # Taller aspect

    for base_name, variants in color_groups.items():

        base_model = variants[0]
        base_color = color_map[base_name]

        y_vals = data[base_model]

        # Zero-shot star
        zero_val = y_vals[0]

        if zero_val is not None:

            if "fmow-mm" in base_model:
                face = "none"   # hollow star
            else:
                face = base_color  # filled star

            plt.plot(
                x[0],
                zero_val,
                marker="*",
                markersize=13,
                linestyle="None",
                color=base_color,
                markerfacecolor=face,
                markeredgewidth=1.2
            )

        # Few-shot curve
        xs = x[1:]
        ys = y_vals[1:]

        plt.plot(
            xs,
            ys,
            linestyle="-",
            marker="o",
            linewidth=2.2,
            color=base_color,
            label=base_model
        )

        # CoOp variants
        for v in variants[1:]:
            y_var = data[v][1:]

            plt.plot(
                xs,
                y_var,
                linestyle=":",
                marker="D",
                linewidth=2.2,
                color=base_color,
                label=v
            )

    # Legend: Zero-shot markers
    filled_star = Line2D(
        [0], [0],
        marker='*',
        linestyle='None',
        markersize=11,
        color='black',
        label='Zero-shot'
    )

    # hollow_star = Line2D(
    #     [0], [0],
    #     marker='*',
    #     linestyle='None',
    #     markersize=11,
    #     markerfacecolor='none',
    #     markeredgecolor='black',
    #     label='Zero-shot (+fmow-mm)'
    # )

    handles, labels = plt.gca().get_legend_handles_labels()
    #handles.extend([filled_star, hollow_star])
    handles.extend([filled_star])

    # Formatting
    plt.xticks(x, shots)
    plt.xlabel("Number of Shots", fontsize=11)
    plt.ylabel("mAP50", fontsize=11)
    plt.title(title, fontsize=13)

    plt.legend(
        handles=handles,
        fontsize=8,
        ncol=1,
        handlelength=3.5,
        handletextpad=0.8,
        columnspacing=1.2,
        frameon=True
    )

    plt.grid(True, alpha=0.25)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_degrade_results(data, title, x_labels, x_label_name, save_path=None):
    x = np.arange(len(x_labels))
    plt.figure(figsize=(9, 6))

    for model, vals in data.items():
        line_ovdsat, = plt.plot(
            x,
            vals["ovdsat"],
            marker="o",
            linestyle="-",
            linewidth=2,
            label=f"{model} ovdsat"
        )
        model_color = line_ovdsat.get_color()

        if "coop" in vals:
            plt.plot(
                x,
                vals["coop"],
                marker="D",
                linestyle="--",
                linewidth=2,
                color=model_color,
                label=f"{model} ovdsat+coop"
            )

        if "zero-shot" in vals:
            plt.plot(
                -0.5,
                vals["zero-shot"],
                marker="*",
                markersize=12,
                linestyle="None",
                color=model_color,
                label=f"{model} zero-shot"
            )

    plt.xticks(x, x_labels)
    plt.xlabel(x_label_name)
    plt.ylabel("mAP50")
    plt.title(title)
    
    ymin, ymax = plt.ylim()
    plt.ylim(ymin, ymax * 1.2)

    plt.legend(fontsize=7, loc="upper right", frameon=True)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    if save_path is not None:
        plt.savefig(save_path, dpi=300, bbox_inches="tight")
        
    plt.show()
```
