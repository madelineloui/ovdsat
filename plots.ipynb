## Few-shot Plots

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

def plot_fewshot(data, title):

    # X axis
    shots = ["0", "1", "3", "5", "10"]
    x = np.array([0, 1, 3, 5, 10])

    # Backbone grouping
    color_groups = {
        "CLIP": ["CLIP", "CLIP + CoOp"],
        "GeoRSCLIP": ["GeoRSCLIP", "GeoRSCLIP + CoOp"],
        "RemoteCLIP": ["RemoteCLIP", "RemoteCLIP + CoOp"],
        "RemoteCLIP + fmow-mm": ["RemoteCLIP + fmow-mm", "RemoteCLIP + fmow-mm + CoOp"],
        "GeoRSCLIP + fmow-mm": ["GeoRSCLIP + fmow-mm", "GeoRSCLIP + fmow-mm + CoOp"],
        "DINOv2": ["DINOv2"],
    }

    # Distinct color assignment (auto, high contrast)
    cmap = plt.get_cmap("tab10")

    color_map = {}
    for i, base_name in enumerate(color_groups.keys()):
        color_map[base_name] = cmap(i)

    # Plot
    plt.figure(figsize=(9, 7))   # Taller aspect

    for base_name, variants in color_groups.items():

        base_model = variants[0]
        base_color = color_map[base_name]

        y_vals = data[base_model]

        # Zero-shot star
        zero_val = y_vals[0]

        if zero_val is not None:

            if "fmow-mm" in base_model:
                face = "none"   # hollow star
            else:
                face = base_color  # filled star

            plt.plot(
                x[0],
                zero_val,
                marker="*",
                markersize=13,
                linestyle="None",
                color=base_color,
                markerfacecolor=face,
                markeredgewidth=1.2
            )

        # Few-shot curve
        xs = x[1:]
        ys = y_vals[1:]

        plt.plot(
            xs,
            ys,
            linestyle="-",
            marker="o",
            linewidth=2.2,
            color=base_color,
            label=base_model
        )

        # CoOp variants
        for v in variants[1:]:
            y_var = data[v][1:]

            plt.plot(
                xs,
                y_var,
                linestyle=":",
                marker="D",
                linewidth=2.2,
                color=base_color,
                label=v
            )

    # Legend: Zero-shot markers
    filled_star = Line2D(
        [0], [0],
        marker='*',
        linestyle='None',
        markersize=11,
        color='black',
        label='Zero-shot'
    )

    hollow_star = Line2D(
        [0], [0],
        marker='*',
        linestyle='None',
        markersize=11,
        markerfacecolor='none',
        markeredgecolor='black',
        label='Zero-shot (+fmow-mm)'
    )

    handles, labels = plt.gca().get_legend_handles_labels()
    handles.extend([filled_star, hollow_star])

    # Formatting
    plt.xticks(x, shots)
    plt.xlabel("Number of Shots", fontsize=11)
    plt.ylabel("mAP50", fontsize=11)
    plt.title(title, fontsize=13)

    plt.legend(
        handles=handles,
        fontsize=8,
        ncol=2,
        handlelength=3.5,
        handletextpad=0.8,
        columnspacing=1.2,
        frameon=True
    )

    plt.grid(True, alpha=0.25)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_degrade_results(data, title, x_labels, x_label_name):
    """
    data format:
    {
      "CLIP": {
          "zero-shot": float,
          "ovdsat": [...],
          "coop":   [...],
      },
      ...
    }
    """

    models = list(data.keys())
    x = np.arange(len(x_labels))

    plt.figure(figsize=(8, 5))

    for model in models:

        # --------
        # ovdsat line (get color)
        # --------
        line_ovdsat, = plt.plot(
            x,
            data[model]["ovdsat"],
            marker="o",
            linestyle="-",
            linewidth=2,
            label=f"{model} ovdsat"
        )

        model_color = line_ovdsat.get_color()

        # --------
        # coop line
        # --------
        plt.plot(
            x,
            data[model]["coop"],
            marker="D",
            linestyle="--",
            linewidth=2,
            color=model_color,
            label=f"{model} ovdsat+coop"
        )

        # --------
        # zero-shot star
        # --------
        if "zero-shot" in data[model]:
            plt.plot(
                -0.5,   # place slightly left of first tick
                data[model]["zero-shot"],
                marker="*",
                markersize=12,
                linestyle="None",
                color=model_color,
                label=f"{model} zero-shot"
            )

    plt.xticks(x, x_labels)
    plt.xlabel(x_label_name)
    plt.ylabel("mAP50")
    plt.title(title)

    plt.legend(handlelength=4, handletextpad=0.8, fontsize=9)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# N regimes
N = [1, 3, 5, 10, 30]

# --- similarity ---
sim_01_no  = [0.0698, 0.0797, 0.1005, 0.1161, 0.1420]
sim_01_aug = [0.0736, 0.0864, 0.1002, 0.1341, 0.1518]

sim_02_no  = [0.0760, 0.1003, 0.1186, 0.1540, 0.1814]
sim_02_aug = [0.0948, 0.1042, 0.1211, 0.1698, 0.1874]

# --- softmax ---
soft_04_no  = [0.0796, 0.0941, 0.1126, 0.1424, 0.1769]
soft_04_aug = [0.0877, 0.1055, 0.1146, 0.1600, 0.1866]

soft_05_no  = [0.0697, 0.0818, 0.1312, 0.1502, 0.1924]
soft_05_aug = [0.0982, 0.1001, 0.1241, 0.1825, 0.1854]

soft_06_no  = [0.0743, 0.0673, 0.1179, 0.1560, 0.1861]
soft_06_aug = [0.0689, 0.0800, 0.1308, 0.1666, 0.1912]


# ----------------------------
# Plot 1  Avg Aug vs No Aug
# ----------------------------
no_all = np.mean([sim_01_no, sim_02_no, soft_04_no, soft_05_no, soft_06_no], axis=0)
aug_all = np.mean([sim_01_aug, sim_02_aug, soft_04_aug, soft_05_aug, soft_06_aug], axis=0)

plt.figure(figsize=(10,7))
plt.plot(N, no_all, marker="o", label="No Aug")
plt.plot(N, aug_all, marker="o", label="Aug")
plt.title("Aug vs No Aug (Avg over thresholds)")
plt.xlabel("N")
plt.ylabel("Score")
plt.legend()
plt.show()


# ----------------------------
# Plot 2  Each threshold
# ----------------------------
plt.figure(figsize=(10,7))

plt.plot(N, sim_01_no,  marker="o", label="sim 0.1 (no aug)")
plt.plot(N, sim_01_aug, marker="o", label="sim 0.1 (aug)")
plt.plot(N, sim_02_no,  marker="o", label="sim 0.2 (no aug)")
plt.plot(N, sim_02_aug, marker="o", label="sim 0.2 (aug)")

plt.plot(N, soft_04_no,  marker="o", label="softmax 0.4 (no aug)")
plt.plot(N, soft_04_aug, marker="o", label="softmax 0.4 (aug)")
plt.plot(N, soft_05_no,  marker="o", label="softmax 0.5 (no aug)")
plt.plot(N, soft_05_aug, marker="o", label="softmax 0.5 (aug)")
plt.plot(N, soft_06_no,  marker="o", label="softmax 0.6 (no aug)")
plt.plot(N, soft_06_aug, marker="o", label="softmax 0.6 (aug)")

plt.title("Similarity + Softmax Thresholds")
plt.xlabel("N")
plt.ylabel("Score")
plt.legend()
plt.show()
```
